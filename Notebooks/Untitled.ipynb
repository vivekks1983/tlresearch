```python
import pandas as pd
import matplotlib.pyplot as plt    
import pandas as pd
from pandas.io.json import json_normalize
from pymongo import MongoClient
import numpy as np
import matplotlib.pyplot as plt
import datetime
from matplotlib.gridspec import GridSpec

client = MongoClient()
db = client.test
collection = db['HMODataFOI']
cursor = list(collection.find({}, {'_id': False}))
df =  pd.DataFrame(list(cursor))
df=df[['Licensed','Unlicensed']].astype(float)
ind = np.arange(26)
width = 0.35
ax = df.plot(kind='bar', title ="HMO Data for FOI",figsize=(20,10),legend=True, fontsize=15)
ax.set_xlabel("Council",fontsize=12)
ax.set_xticks(ind + width / 2)
ax.set_ylabel("Licensed/Unlicensed Houses",fontsize=12)
plt.show()
```
